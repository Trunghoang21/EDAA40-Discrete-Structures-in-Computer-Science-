This exercise is about one of the most famous proofs in mathematics and the first proper mathematical proof that many students see, namely Euclid's proof that there are infinitely many prime numbers. There are many variations of it, and in particular it need not be phrased as a proof by contradiction as we do here. The [Wikipedia page](https://en.wikipedia.org/wiki/Euclid%27s_theorem) goes into more detail about this and might be an interesting read after completing the exercise.

---
**Theorem:** There are infinitely many prime numbers. 

*Proof:* Toward a contradiction, suppose that there are finitely many, say $N$, prime numbers, and denote them in ascending order by $p_1, \ldots, p_N$. Let $p$ be the product of all these primes: $p = p_1\cdot\ldots \cdot p_N$. Write $q$ to denote the number $p+1$. We prove that $q$ is prime, reasoning by cases. 

- If $q$ is prime, then since $q$ is not one of $p_1, \ldots, p_N$ we get a contradiction with the fact that $p_1, \ldots, p_N$ was a list of all prime numbers.
- If $q$ is not prime, then $q$ has a prime factor $r$. Since $p$ is the product of all prime numbers, it must hold that $r$ is also a prime factor of $p$. Then $r$ is also a prime factor of $q-p = p + 1 - p = 1$. Since no prime number divides $1$, the number $r$ cannot be one of the numbers $p_i$, again contradicting that $p_1, \ldots, p_N$ is a list of all prime numbers.
---

Euclid's proof crucially relies on numbers of the form $p_1 \cdot \ldots \cdot p_N + 1$ where $p_1, \ldots, p_N$ are the first $N$ prime numbers. Numbers of this form are called _Euclid numbers_, and your task in this exercise is to investigate the properties of Euclid numbers using programming in Python. **Task 2:** Write a Python function called `euclid_number` that receives an integer $i$ as input and outputs the $i$-th Euclid number using the list `primes` that you defined. 

Finally, print the first four Euclid numbers. Are they prime numbers?

```python
# If you get an error saying ''primes'' is not defined, run the code block above.
def euclid_number(i):
    re = 1
    for idx in range (i):
        re *= primes[idx]
    return re + 1

for i in range(10):
    print(euclid_number(i))
```

**Task 3:** Write a Python function called `is_prime` that receives a number $n$ as input and outputs True if the number is prime and False otherwise. Try it out on the numbers $4, 6, 11111111$, and $7$. 

```python

def is_prime(n):
    if n <= 1:
        return False
    else:
        for i in range(2, n):
            if n % i == 0:
                return False
    return True
    
print(is_prime(4))
print(is_prime(6))
print(is_prime(11111111))
print(is_prime(7))
```
